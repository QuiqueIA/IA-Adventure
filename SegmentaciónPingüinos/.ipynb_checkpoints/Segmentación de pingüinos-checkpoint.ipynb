```python
# Importación de librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Leemos el dataset
pingus = pd.read_csv("penguins.csv")
print(pingus.info())
print(pingus.describe())
```

```python
pingus
```

```python
# Análisis estadístico gráfico.
sns.pairplot(pingus)
```

```python
# Asignación de valores a los NA
pingus["culmen_length_mm"].fillna(pingus["culmen_length_mm"].mean(),inplace=True)
pingus["culmen_depth_mm"].fillna(pingus["culmen_depth_mm"].mean(),inplace=True)
pingus["flipper_length_mm"].fillna(pingus["flipper_length_mm"].mean(),inplace=True)
pingus["body_mass_g"].fillna(pingus["body_mass_g"].mean(),inplace=True)
```

```python
# Eliminar valores inválidos de SEX
pingus = pingus[pingus["sex"]!= "."]
pingus.isna().sum()
```

```python
# Convertir a numérico la variable categórica y estandarizar.
pingus = pd.get_dummies(pingus,columns=["sex"],drop_first=True)
pingus_escaled = StandardScaler().fit_transform(pingus)
```

```python
# Aplicamos PCA para reducir dimensionalidad
pca = PCA(n_components=None)
df_pca = pca.fit(pingus_escaled)
n_components = sum(df_pca.explained_variance_ratio_>0.1)
pca = PCA(n_components=n_components)
df_pca = pca.fit_transform(pingus_escaled)
print(n_components)
```

```python
# Cálculo y selección del mejor número de clústers para K-Means. Método del codo
k_range = range(1,11)
def kmeans_inertias(num_clusters,df_pca):
    inertias = []
    for c in k_range:
        km = KMeans(n_clusters=c,random_state=0)
        km.fit(df_pca)
        inertias.append(km.inertia_)
    return inertias

inertias = kmeans_inertias(k_range,df_pca)
print(inertias)

sns.lineplot(x=k_range,y=inertias,marker='o')
```

```python
# Creación del modelo
kmeans=KMeans(n_clusters=5,random_state=42)
kmeans.fit(df_pca)
```

```python
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=kmeans.labels_, cmap="viridis")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title(f"K-means Clustering (K={5})")
plt.show()
```
