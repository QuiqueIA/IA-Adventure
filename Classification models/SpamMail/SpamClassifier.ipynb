```python
# Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import spacy
import nltk
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
from collections import Counter
import seaborn as sns
```

```python
# Lectura del dataframe
df = pd.read_csv("spam.csv")
df.head()
```

```python
# Análisis lingüístico
spam_words = ' '.join(df[df['Category'] == 'spam']['Message']).split()
ham_words = ' '.join(df[df['Category'] == 'ham']['Message']).split()

spam_word_counts = Counter(spam_words)
ham_word_counts = Counter(ham_words)

spam_common_words = spam_word_counts.most_common(10)
ham_common_words = ham_word_counts.most_common(10)

spam_df = pd.DataFrame(spam_common_words, columns=['Word', 'Count'])
ham_df = pd.DataFrame(ham_common_words, columns=['Word', 'Count'])

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.barplot(x='Count', y='Word', data=spam_df, palette='Reds')
plt.title('Most Common Words in Spam Messages')

plt.subplot(1, 2, 2)
sns.barplot(x='Count', y='Word', data=ham_df, palette='Blues')
plt.title('Most Common Words in Ham Messages')

plt.tight_layout()
plt.show()
```

```python
# Preprocesamiento de mensajes y NLP
# 1) Eliminación de datos nulos
df = df.dropna()

# 2) Eliminación de duplicidades.
df = df.drop_duplicates()

# 3) Procesamiento de los mensajes
nlp = spacy.load('en_core_web_sm')

def clean_text(text):
    # Procesar el texto con spaCy
    doc = nlp(text)
    
    # Filtrar stopwords, puntuaciones, y convertir a minúsculas
    cleaned_tokens = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct]
    
    # Unir los tokens en un solo string
    cleaned_text = " ".join(cleaned_tokens)
    
    return cleaned_text

df['Processed_Message'] = df['Message'].apply(clean_text)
df.head(5)
```

```python
# Preparación de datos de entrenamiento.
# Crear el LabelEncoder
label_encoder = LabelEncoder()
categories = df["Category"]

# Ajustar y transformar las etiquetas
encoded_labels = label_encoder.fit_transform(categories)

# Mostrar el resultado
print("Etiquetas originales:", categories)
print("Etiquetas codificadas:", encoded_labels)

# Extraer características del texto procesado
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df["Processed_Message"]).toarray()
Y = encoded_labels

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# Creación de los modelos
%time
models = {
    "AdaBoost": AdaBoostClassifier(algorithm='SAMME'),  # Corregido warning
    "Random Forest": RandomForestClassifier(),
    "Naive Bayes": GaussianNB(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(max_iter=1000),
    "SVM": SVC(probability=True),
    "Decision Tree": DecisionTreeClassifier()
}

for k, v in models.items():
    print ('CREANDO MODELO: {clas}'.format(clas=k))
    v.fit(X_train, y_train)   
```

```python
# Evaluación de los modelos
def evaluate_model(model, X_test, y_test, model_name):
    # Predicciones
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]  # Probabilidades para la clase positiva (ROC)

    # Métricas principales
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"Modelo: {model_name}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print("-" * 50)

    # Matriz de confusión
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['No Spam', 'Spam'], yticklabels=['No Spam', 'Spam'])
    plt.title(f"Matriz de Confusión - {model_name}")
    plt.xlabel("Predicción")
    plt.ylabel("Real")
    plt.show()

    # Curva ROC
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Curva ROC - {model_name}')
    plt.legend(loc="lower right")
    plt.show()

# Evaluar cada modelo
for model_name, model in models.items():
    evaluate_model(model, X_test, y_test, model_name)  # Evaluar el modelo
```
