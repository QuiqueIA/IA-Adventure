```python
# Importamos las librerías.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Leemos dataset y preparamos datos.
stocks = pd.read_csv("Microsoft_Stock.csv",parse_dates=["Date"],index_col="Date")
# Usamos la columna close
stocks = stocks[["Close"]]
stocks = stocks.asfreq("B")
stocks = stocks.fillna(method="ffill")
stocks
```

```python
plt.figure(figsize=(12,6))
plt.plot(stocks,label="Precio de Cierre Real")
plt.title("Serie temporal del precio de cierre de Microsoft")
plt.legend()
plt.show()
```

```python
# Creación del modelo de red neuronal recurrente
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(stocks)

# Creación de secuencias de entrenamiento
def create_sequences(sequence,n_steps):
    X,y=[],[]
    for i in range(len(sequence)):
        end_ix= i+n_steps
        if end_ix>len(sequence)-1:
            break
        seq_x,seq_y = sequence[i:end_ix],sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X),np.array(y)

n_steps = 60
X,y = create_sequences(scaled_data,n_steps)

split = int(0.8*len(X))
X_train,X_test = X[:split],X[split:]
y_train,y_test = y[:split],y[split:]

# Modelo LSTM
model_lstm = Sequential([
    LSTM(100,activation="relu",input_shape=(n_steps,1)),
    Dense(1)
])
model_lstm.compile(optimizer="adam",loss="mse")
history = model_lstm.fit(X_train,y_train,epochs=200)
model_lstm.save_weights("weights.weights.h5")

plt.figure(figsize=(10,6))
plt.plot(history.history["loss"],label="Pérdida en entrenamiento")
plt.title("Proceso de Aprendizaje del modelo LSTM")
plt.xlabel("Épocas")
plt.ylabel("Pérdida (MSE)")
plt.legend()
plt.show()
```

```python
# Evaluamos el modelo con predicciones pasadas
model_lstm.load_weights("weights.weights.h5")

# Predecir para todo el conjunto de datos
all_predictions = model_lstm.predict(X)
all_predictions = scaler.inverse_transform(all_predictions)  # Invertir la normalización

# Graficar los valores reales y las predicciones
plt.figure(figsize=(14, 7))
plt.plot(stocks.index, stocks["Close"], label="Valores Reales", color="blue")
plt.plot(stocks.index[n_steps:], all_predictions, label="Predicciones del Modelo", color="red", linestyle="--")
plt.title("Comportamiento del Modelo LSTM en Todo el Conjunto de Datos")
plt.xlabel("Fecha")
plt.ylabel("Precio de Cierre")
plt.legend()
plt.grid()
plt.show()

# Calcular el error en las predicciones futuras (si tienes y_true_future)
# y_true_future debe ser un array con los valores reales correspondientes a las predicciones futuras
if 'y_true_future' in locals():
    mse = mean_squared_error(y_true_future, future_predictions)
    mae = mean_absolute_error(y_true_future, future_predictions)
    print(f"Error Cuadrático Medio (MSE): {mse}")
    print(f"Error Absoluto Medio (MAE): {mae}")
else:
    print("No se tienen valores reales futuros para calcular el error.")
```

```python
# Evaluamos el modelo con predicciones futuras
model_lstm.load_weights("weights.weights.h5")

# Tomar los últimos n_steps valores de los datos de prueba
last_sequence = X_test[-1]  # Última secuencia de los datos de prueba
future_predictions = []  # Lista para almacenar las predicciones futuras
n_future = 30  # Número de predicciones futuras que queremos hacer

# Hacer predicciones a futuro
for _ in range(n_future):
    # Predecir el siguiente valor
    next_prediction = model_lstm.predict(last_sequence.reshape(1, n_steps, 1))
    
    # Guardar la predicción
    future_predictions.append(next_prediction[0, 0])
    
    # Actualizar la secuencia para la siguiente predicción
    last_sequence = np.append(last_sequence[1:], next_prediction)

# Invertir la normalización de las predicciones
future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Generar fechas futuras
last_date = stocks.index[-1]  # Última fecha en los datos originales
future_dates = pd.date_range(last_date, periods=n_future + 1, freq="B")[1:]  # Fechas futuras

# Graficar los valores reales y las predicciones futuras
plt.figure(figsize=(14, 7))
plt.plot(stocks.index[-100:], stocks["Close"].values[-100:], label="Valores Reales", color="blue")
plt.plot(future_dates, future_predictions, label="Predicciones Futuras", color="red", linestyle="--")
plt.title("Predicciones Futuras del Precio de Cierre con LSTM")
plt.xlabel("Fecha")
plt.ylabel("Precio de Cierre")
plt.legend()
plt.grid()
plt.show()
```
