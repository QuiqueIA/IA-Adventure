```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import differential_evolution
```

```python
# 1️⃣ Cargar el dataset
file_path = "Microsoft_Stock.csv"
data = pd.read_csv(file_path)

# 2️⃣ Preprocesamiento de datos
data["Date"] = pd.to_datetime(data["Date"])  # Convertir la fecha a tipo datetime
data.set_index("Date", inplace=True)  # Usar la fecha como índice
prices = data["Close"].values  # Extraer precios de cierre
```

```python
# 3️⃣ Definir la función objetivo para optimizar la estrategia de trading
def trading_strategy(params):
    short_window, long_window = int(params[0]), int(params[1])
    
    if short_window >= long_window:
        return np.inf  # Penalización si el periodo corto es mayor o igual al largo
    
    short_ma = pd.Series(prices).rolling(window=short_window).mean()
    long_ma = pd.Series(prices).rolling(window=long_window).mean()
    
    signals = np.where(short_ma > long_ma, 1, -1)  # Señales de compra (1) y venta (-1)
    returns = np.diff(prices) * signals[:-1]  # Calcular retornos
    return -np.sum(returns)  # Negativo porque scipy minimiza la función objetivo
```

```python
# 4️⃣ Ejecutar el algoritmo genético
bounds = [(5, 50), (50, 200)]  # Límites para short_window y long_window
result = differential_evolution(trading_strategy, bounds, strategy="best1bin", maxiter=100, popsize=20)
```

```python
# 5️⃣ Obtener parámetros óptimos
opt_short_window, opt_long_window = int(result.x[0]), int(result.x[1])
print(f"Parámetros óptimos: Short Window = {opt_short_window}, Long Window = {opt_long_window}")

# 6️⃣ Generar medias móviles con los mejores parámetros
data["Short_MA"] = data["Close"].rolling(window=opt_short_window).mean()
data["Long_MA"] = data["Close"].rolling(window=opt_long_window).mean()
```

```python
# 7️⃣ Generar señales de compra/venta
signals = np.where(data["Short_MA"] > data["Long_MA"], 1, -1)
data["Signal"] = signals
buy_signals = data[data["Signal"] == 1]
sell_signals = data[data["Signal"] == -1]
```

```python
# 8️⃣ Visualizar los resultados
plt.figure(figsize=(12, 6))
plt.plot(data.index, data["Close"], label="Precio de Cierre", color="black")
plt.plot(data.index, data["Short_MA"], label=f"Media Móvil ({opt_short_window} días)", color="blue", linestyle="dashed")
plt.plot(data.index, data["Long_MA"], label=f"Media Móvil ({opt_long_window} días)", color="red", linestyle="dashed")
plt.scatter(buy_signals.index, buy_signals["Close"], label="Compra", marker="^", color="green", alpha=1)
plt.scatter(sell_signals.index, sell_signals["Close"], label="Venta", marker="v", color="red", alpha=1)

plt.title("Optimización de Trading con Algoritmo Genético")
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.legend()
plt.grid()
plt.show()
```
