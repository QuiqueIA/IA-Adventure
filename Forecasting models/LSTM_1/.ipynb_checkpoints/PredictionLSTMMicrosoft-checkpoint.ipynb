```python
# Importamos las librer√≠as.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Leemos dataset y preparamos datos.
stocks = pd.read_csv("Microsoft_Stock.csv", parse_dates=["Date"], index_col="Date")
stocks = stocks[["Close"]].asfreq("B").fillna(method="ffill")

# Visualizaci√≥n de la serie temporal
plt.figure(figsize=(12,6))
plt.plot(stocks, label="Precio de Cierre Real")
plt.title("Serie temporal del precio de cierre de Microsoft")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(stocks,label="Precio de Cierre Real")
plt.title("Serie temporal del precio de cierre de Microsoft")
plt.legend()
plt.show()
```

```python
# Normalizaci√≥n de datos
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(stocks)

# Dividir en entrenamiento y datos futuros (80% - 20%)
split_ratio = 0.8
split_index = int(len(scaled_data) * split_ratio)
scaled_data_train = scaled_data[:split_index]
scaled_data_future = scaled_data[split_index:]
```

```python
# Funci√≥n para crear secuencias
def create_sequences(sequence, n_steps):
    X, y = [], []
    for i in range(len(sequence)):
        end_ix = i + n_steps
        if end_ix > len(sequence)-1:
            break
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
n_steps = 60
X, y = create_sequences(scaled_data_train, n_steps)  # Solo usar datos de entrenamiento

# Dividir en entrenamiento y validaci√≥n
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
# Modelo LSTM
model_lstm = Sequential([
    LSTM(100, activation="relu", input_shape=(n_steps, 1)),
    Dense(1)
])
model_lstm.compile(optimizer="adam", loss="mse")
history = model_lstm.fit(X_train, y_train, epochs=200, validation_data=(X_test, y_test))

# Guardar pesos
model_lstm.save_weights("weights.weights.h5")
```

```python
# Curva de aprendizaje
plt.figure(figsize=(10,6))
plt.plot(history.history["loss"], label="Entrenamiento")
plt.plot(history.history["val_loss"], label="Validaci√≥n")
plt.title("Curva de aprendizaje del modelo LSTM")
plt.xlabel("√âpocas")
plt.ylabel("P√©rdida (MSE)")
plt.legend()
plt.show()
```

```python
# Predicciones futuras
model_lstm.load_weights("weights.weights.h5")

# Obtener √∫ltima secuencia de los datos de entrenamiento
last_sequence = scaled_data_train[-n_steps:].reshape(1, n_steps, 1)
future_predictions_scaled = []
```

```python
# Generar predicciones recursivas
for _ in range(len(scaled_data_future)):
    pred = model_lstm.predict(last_sequence, verbose=0)
    future_predictions_scaled.append(pred[0,0])
    last_sequence = np.concatenate([last_sequence[:,1:,:], pred.reshape(1,1,1)], axis=1)
    # Invertir normalizaci√≥n
future_predictions = scaler.inverse_transform(np.array(future_predictions_scaled).reshape(-1,1))
y_true_future = scaler.inverse_transform(scaled_data_future)
```

```python
# C√°lculo de m√©tricas
mse = mean_squared_error(y_true_future, future_predictions)
mae = mean_absolute_error(y_true_future, future_predictions)
print(f"\nError en predicciones futuras:")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
# Visualizaci√≥n de resultados
plt.figure(figsize=(14,7))

# Datos reales
plt.plot(stocks.index, stocks["Close"], label="Real", color="blue")

# Predicciones en datos de entrenamiento/validaci√≥n
train_predictions = scaler.inverse_transform(model_lstm.predict(X_train))
val_predictions = scaler.inverse_transform(model_lstm.predict(X_test))

train_dates = stocks.index[n_steps : n_steps + len(X_train)]
val_dates = stocks.index[n_steps + len(X_train) : n_steps + len(X_train) + len(X_test)]

plt.plot(train_dates, train_predictions, color="green", linestyle="--", label="Entrenamiento")
plt.plot(val_dates, val_predictions, color="orange", linestyle="--", label="Validaci√≥n")

# Predicciones futuras
future_dates = stocks.index[split_index : split_index + len(scaled_data_future)]
plt.plot(future_dates, future_predictions, color="red", linestyle="--", label="Predicciones Futuras")

plt.title("Comparaci√≥n de predicciones")
plt.xlabel("Fecha")
plt.ylabel("Precio de Cierre")
plt.legend()
plt.grid()
plt.show()
```
