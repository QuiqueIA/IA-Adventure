```python
# Importamos las librer√≠as.
import gym
import numpy as np
import matplotlib.pyplot as plt
env = gym.make("FrozenLake-v1", desc=None, map_name="4x4", is_slippery=False)
```

```python
# Definici√≥n del agente Q-Learning
alpha = 0.1       # Tasa de aprendizaje
gamma = 0.99      # Factor de descuento
epsilon = 1.0     # Exploraci√≥n inicial
epsilon_decay = 0.995  # Decaimiento de exploraci√≥n
MIN_EPSILON = 0.01  # Exploraci√≥n m√≠nima

# Q-Table (representa el conocimiento del agente)
q_table = np.zeros([env.observation_space.n, env.action_space.n])

# Entrenamiento mejorado con registro de recompensas
def train_agent(env, episodes=10000, max_steps=100):
    global epsilon
    rewards = []
    moving_avg = []
    
    for episode in range(episodes):
        state, _ = env.reset()  # Cambio aqu√≠
        total_reward = 0
        done = False
        step = 0
        
        while not done and step < max_steps:
            if np.random.uniform() < epsilon:
                action = env.action_space.sample()
            else:
                action = np.argmax(q_table[state])
            
            # Cambio cr√≠tico aqu√≠ (manejo de nueva API)
            next_state, reward, terminated, truncated, info = env.step(action)  # Desempaquetado correcto
            done = terminated or truncated  # Nueva l√≥gica de terminaci√≥n
            
            # Resto del c√≥digo igual...
            old_value = q_table[state, action]
            next_max = np.max(q_table[next_state])
            new_value = (1 - alpha) * old_value + alpha * (reward + gamma * next_max)
            q_table[state, action] = new_value
            
            state = next_state
            total_reward += reward
            step += 1
        
        epsilon = max(MIN_EPSILON, epsilon * epsilon_decay)
        rewards.append(total_reward)
        moving_avg.append(np.mean(rewards[-100:]))
    
    return rewards, moving_avg

# Entrenar el agente y obtener m√©tricas
episodes = 10000
rewards, moving_avg = train_agent(env, episodes=episodes)

# 1. Graficar el aprendizaje
plt.figure(figsize=(12, 6))
plt.plot(rewards, alpha=0.4, label='Recompensa por episodio')
plt.plot(moving_avg, linewidth=2, label='Media m√≥vil (100 episodios)')
plt.xlabel('Episodios')
plt.ylabel('Recompensa')
plt.title('Progreso del aprendizaje')
plt.legend()
plt.savefig('progreso_aprendizaje.png')  # Guardar gr√°fico
plt.show()

# 2. Guardar el modelo entrenado
np.save('q_table.npy', q_table)
print("Modelo guardado exitosamente como q_table.npy")

# Evaluaci√≥n corregida (usando la posici√≥n real de la meta)
exitos = 0
num_evaluaciones = 100

# Calcular estado objetivo usando el grid
grid = env.unwrapped.desc.flatten()  # Obtener representaci√≥n del mapa
goal_state = np.where(grid == b'G')[0][0]  # Encontrar √≠ndice de la meta

for _ in range(num_evaluaciones):
    state, _ = env.reset()
    state = int(state)
    done = False
    
    while not done:
        action = np.argmax(q_table[state])
        next_state, _, terminated, truncated, _ = env.step(action)
        done = terminated or truncated
        state = int(next_state)
    
    if state == goal_state:  # Comparar con posici√≥n real de la meta
        exitos += 1

print(f"\nTasa de √©xito: {exitos/num_evaluaciones:.2%}")
print("Mapa del entorno:")
print(env.unwrapped.desc)
```
